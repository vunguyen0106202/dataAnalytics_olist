```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
import pandas as pd
df=pd.read_csv('data.csv')
df['price'] = pd.to_numeric(df['price'], errors='coerce')
df['freight_value'] = pd.to_numeric(df['freight_value'], errors='coerce')
df['sale'] = df['price'] + df['freight_value']
features = ['price', 'freight_value', 'review_score']
target = 'order_status'

# Chọn các cột có mặt số
X =df[features]
y = df[target]

# Chuyển đổi các cột mục tiêu thành số
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(y)

# Chia dữ liệu thành tập huấn luyện và tập kiểm tra
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Tiền xử lý dữ liệu (chuẩn hóa)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
# Khởi tạo mô hình
model = RandomForestClassifier(n_estimators=100, random_state=42)

# Huấn luyện mô hình
model.fit(X_train, y_train)

# Dự đoán trên tập kiểm tra
y_pred = model.predict(X_test)


# Chuyển đổi dự đoán về dạng văn bản (nếu cần)
y_pred_labels = label_encoder.inverse_transform(y_pred)

# Tạo DataFrame chứa các dự đoán
results = pd.DataFrame({
    'Actual': label_encoder.inverse_transform(y_test),
    'Predicted': y_pred_labels
})

# Hiển thị 20 dự đoán đầu tiên
print(results.head(20))
# Đánh giá mô hình
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```


Confusion Matrix (Ma trận nhầm lẫn) là một công cụ hữu ích trong việc đánh giá hiệu suất của mô hình phân loại. Nó giúp bạn hiểu cách mô hình của bạn hoạt động với từng lớp trong tập dữ liệu. Dưới đây là cách giải thích chi tiết về confusion matrix:

Cấu trúc của Confusion Matrix
Confusion matrix là một ma trận vuông, trong đó mỗi hàng đại diện cho lớp thực tế và mỗi cột đại diện cho lớp dự đoán của mô hình.

Cách Đọc Confusion Matrix
Chỉ số chính xác cho lớp i: ma trận[i][i] cho biết số lượng mẫu của lớp i mà mô hình đã dự đoán đúng.
Số lỗi cho lớp i: Tổng các giá trị trong hàng i trừ đi giá trị chính xác (sai sót được dự đoán thành các lớp khác).
Tổng số mẫu cho lớp i: Tổng các giá trị trong hàng i, cho biết tổng số mẫu thực tế thuộc lớp i.
Phân Tích
Precision cho lớp i: Tỷ lệ dự đoán đúng cho lớp i so với tổng số dự đoán là lớp i.
Recall cho lớp i: Tỷ lệ dự đoán đúng cho lớp i so với tổng số mẫu thực tế thuộc lớp i.
Confusion matrix giúp bạn phân tích hiệu suất mô hình phân loại của mình, đặc biệt là khi bạn có nhiều lớp và cần biết mô hình có xu hướng dự đoán sai cho lớp nào. Nếu có bất kỳ câu hỏi nào khác hoặc cần thêm thông tin, đừng ngần ngại hỏi!

Kết quả này là bảng đánh giá mô hình phân loại của bạn, được tạo ra bằng cách sử dụng các chỉ số như precision, recall, f1-score và support. Đây là cách bạn có thể hiểu các thông số trong bảng:

1. **Confusion Matrix**: 
   - Đoạn ma trận đầu tiên là confusion matrix, cho biết số lượng dự đoán đúng và sai cho từng lớp. Mỗi hàng tương ứng với lớp thực tế và mỗi cột tương ứng với lớp dự đoán.

2. **Classification Report**:
   - **Precision**: Tỷ lệ dự đoán chính xác trong số các dự đoán là lớp đó (ví dụ, precision của lớp 0 là 0.25 có nghĩa là 25% của các dự đoán lớp 0 là chính xác).
   - **Recall**: Tỷ lệ đúng các mẫu thuộc lớp đó được dự đoán đúng (ví dụ, recall của lớp 1 là 0.25 có nghĩa là 25% của các mẫu thực sự thuộc lớp 1 được dự đoán đúng).
   - **F1-Score**: Trung bình hài hòa của precision và recall (F1-score của lớp 0 là 0.40).
   - **Support**: Số lượng mẫu thực tế thuộc lớp đó trong tập kiểm tra.

3. **Accuracy**: Tỷ lệ chính xác tổng thể của mô hình (0.98 có nghĩa là 98% các dự đoán là chính xác).

4. **Macro Average**: Trung bình của các chỉ số (precision, recall, f1-score) trên tất cả các lớp, không cân nhắc đến tần suất của từng lớp.

5. **Weighted Average**: Trung bình các chỉ số trên tất cả các lớp, cân nhắc đến số lượng mẫu của từng lớp.
